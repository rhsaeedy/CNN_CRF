```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import matplotlib.cm as cm
%matplotlib inline
import scipy.misc
from PIL import Image
import scipy.io
import os


# Make sure that caffe is on the python path:
caffe_root = '../../'  # this file is expected to be in {caffe_root}/examples/hed/
import sys
sys.path.insert(0, caffe_root + 'python')

import caffe
```

```python
data_root = '/home/reza_rec/crfasrnn/caffe/examples/Gambier/input/'
with open(data_root + 'images.txt') as f:
    test_lst = f.readlines()
    
test_lst = [data_root + x.strip() for x in test_lst]
```

```python
im_lst = []
for i in range(0, len(test_lst)):
    im = Image.open(test_lst[i])
    in_ = np.array(im, dtype=np.float32)
    in_ = in_[:,:,::-1]
    in_ -= np.array((149.79, 149.79, 149.79))
    im_lst.append(in_)
```

```python
#Visualization
def plot_single_scale(scale_lst, size):
    pylab.rcParams['figure.figsize'] = size, size/2
    
    plt.figure()
    for i in range(0, len(scale_lst)):
        s=plt.subplot(1,6,i+1)
        plt.imshow(scale_lst[i], cmap = cm.Greys_r)
        s.set_xticklabels([])
        s.set_yticklabels([])
        s.yaxis.set_ticks_position('none')
        s.xaxis.set_ticks_position('none')
    plt.tight_layout()
```

```python
idx = 5

in_ = im_lst[idx]
in_ = cv2.resize(in_, dsize=(512, 512), interpolation=cv2.INTER_CUBIC)
in_ = in_.transpose((2,0,1))
#remove the following two lines if testing with cpu

# load net
model_root = './models/'
train_root = './train/'
pretrain_root = './pretrain/'
#net = caffe.Net(model_root + 'gamb_train_py.prototxt', model_root + 'crfrnn_gamb_iter_2403.caffemodel', caffe.TEST)

net = caffe.Net(pretrain_root + 'convnet_deploy.prototxt', model_root + 'convnet_iter_662.caffemodel', caffe.TEST)
# shape for input (data blob is N x C x H x W), set data
net.blobs['Data'].reshape(1, *in_.shape)
net.blobs['Data'].data[...] = in_
# run net and take argmax for prediction
net.forward()
out1 = net.blobs['conv1_1'].data[0][0,:,:]#.astype('uint8')
out2 = net.blobs['conv1_2'].data[0][0,:,:]#.astype('uint8')
out3 = net.blobs['conv2_1'].data[0][0,:,:]#.astype('uint8')
out4 = net.blobs['fc6'].data[0][0,:,:]#.astype('uint8')
out5 = net.blobs['bigscore'].data[0][0,:,:]#.astype('uint8')
out6 = net.blobs['score-pool3c'].data[0][0,:,:]
pred = net.blobs['score4'].data[0][0,:,:]#.astype('uint8')

scale_lst = [pred]
plot_single_scale(scale_lst, 10)
scale_lst = [out1, out2, out3, out4, out5, out6]
plot_single_scale(scale_lst, 10)


#test = pred - out5
#scale_lst = [test]
#plot_single_scale(scale_lst, 10)
```
